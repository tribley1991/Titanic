### Data Wrangling

To fill missing ages, there are several potential methods. 

First, "Pclass" and "Sex" can be used to group the series "age". The median value of each sub-group canis used to fill the missing data in corresponding group.

```python
age_grouped = df.groupby(['Pclass','Sex'])['Age']
```

```python
age_grouped.median()
```

Second, "Parch" and "SibSp" can be used to group passengers. The assumption is that people at similar age will have similar number of siblings and parents/children. However, the group might be all unknown, such as (8,2) group.

```python
age_grouped2 = df.groupby(['SibSp','Parch'])['Age']
```

```python
age_grouped2.agg(['median','size'])
```

Third, "SibSp" and "Parch" can be used to imply age. When "Parch">2, the passenger is obviously a child. When "Parch"<=2, the passenger might be a child or adult.

### Identify Important Factors

#### Sex

To determine the factors which will affect survival, the first guess is "Sex". During evacuation, women had priority to safety boat over men which should affect survival.

```python
sex_grouped = pd.crosstab(df.Sex,df.Survived, margins=True)
```

```python
sex_grouped
```

```python
sex_grouped[0] = sex_grouped[0]/sex_grouped['All']
sex_grouped[1] = sex_grouped[1]/sex_grouped['All']
```

```python
sex_grouped.ix[['female','male'],[0,1]].plot(kind='bar', stacked = True)
```

Higher proportion of female survived the accident than men did. "Sex" should definitely be included for modeling.

#### Age

```python
df['Age'].hist(bins=20,alpha=0.5)
```

```python
fig, axes = plt.subplots(2,1)
df[df['Survived']==0]['Age'].hist(ax=axes[0],bins=20,alpha=0.5)
axes[0].set_title('Age distribution for victims')
axes[0].set_xlim([0,85])
df[df['Survived']==1]['Age'].hist(ax=axes[1],bins=20,alpha=0.5)
axes[1].set_title('Age distribution for survivors')
axes[1].set_xlim([0,85])
```

This demonstrates that childern are more likely to survive than old people. As a result, "age" should be taken into consideration during modeling process.

#### Class

```python
class_grouped = pd.crosstab(df.Pclass,df.Survived,margins=True)
```

```python
class_grouped
```

```python
class_grouped[0]=class_grouped[0]/class_grouped['All']
class_grouped[1]=class_grouped[1]/class_grouped['All']
```

```python
class_grouped.ix[0:3,0:2].plot(kind='bar',stacked=True)
```

As we can see, higher proportion of people survived from higher level classes. Thus, "class" is an important factor for predicting survival.
